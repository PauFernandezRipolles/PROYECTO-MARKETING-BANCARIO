## 3. Miro poblaciones, con igual número de llamadas (campaign) para la campaña actual y con deposito=1 o 0

```python
filter_campaign_inf_10=df_BANK_vista_marketing['campaign']<=10
```

Podemos ajustar todas las distribuciones positivas a una función Gamma, hasta campaign=6 según vemos en las distribuciones

```python
g=sns.FacetGrid(df_BANK_vista_marketing[filter_campaign_inf_10], row="campaign",  col="deposit",palette="crest")
g.map_dataframe(sns.histplot, x="duration",color="green")
```

La estrategia será para cada distribución Gamma, con campaña y depósito, hacer su curva Gamma. En los valores de la distribución con deposito=0, que la probabilidad de que pase
un tiempo superior sea una probabilidad muy baja, será el valor que sí nos puede servir para la distribución con depósito=1

**Cálculo un zoom de la función de distribución, para campaign=1 y deposit=0**

```python
filter_campaign_n_deposit_yn=(df_BANK_vista_marketing['campaign']==1) & (df_BANK_vista_marketing['deposit']==0)
#plot=sns.displot(data=df_BANK_vista_marketing[filter_campaign_n_deposit_no],x="duration",bins=20,height=5, aspect=2)
#plot.set_axis_labels('duración', 'frecuencia')

# Graficamos el histograma con seaborn displot
sns.displot(data=df_BANK_vista_marketing[filter_campaign_n_deposit_yn],x="duration", kde=False, stat='density')
```

Estimo los parámetros de la distribución Gamma

```python
from scipy.stats import gamma

# Datos de ejemplo
data = df_BANK_vista_marketing[filter_campaign_n_deposit_yn]['duration']

# Ajuste de la distribución gamma
params = gamma.fit(data,floc=0)

# Los parámetros ajustados son:
k, loc, theta = params
print(params)
```

Visualizo la curva de la función Gamma, ya que así compruebo que se ajusta a la distribución anterior

```python
x = np.linspace(min(data), max(data), 100)
pdf_gamma = gamma.pdf(x, a=k, loc=loc, scale=theta)

#plot=sns.displot(data=df_BANK_vista_marketing[filter_campaign_n_deposit_no],x="duration",bins=20,height=5, aspect=2)
#plot.set_axis_labels('duración', 'frecuencia')

plt.plot(x, pdf_gamma, 'r-', lw=2, label=f'Gamma fit (k={k:.2f}, θ={theta:.2f})')
plt.legend()


plt.show()
```

Buscamos que duración de llamada a clientes,**que los llaman una única vez en la misma campaña**, que sea mayor de un cierto valor,  y que al final no contraten el depósito, y que esto pase con una probabilidad inferior a 0,01.

```python
p_valor_cola = 0.01

# Calculamos el percentil para la cola superior
# 1 - p_valor_cola da la probabilidad acumulada hasta el valor x 
probabilidad_acumulada = 1 - p_valor_cola

# Usamos la función ppf para encontrar el valor de x correspondiente a esa probabilidad acumulada
x_valor = gamma.ppf(probabilidad_acumulada, a=k, loc=loc, scale=theta)

x_valor
```

La probabilidad que un cliente al que llamen una única vez en la campaña actual y no contrate el depósito y que la llamada pueda durar más de **821,68 segundos**, es inferior a 0,01.

Visualizo la curva de la función Gamma, ya que así compruebo que se ajusta a la distribución anterior

```python
x = np.linspace(min(data), max(data), 100)
pdf_gamma = gamma.pdf(x, a=k, loc=loc, scale=theta)

#plot=sns.displot(data=df_BANK_vista_marketing[filter_campaign_n_deposit_no],x="duration",bins=20,height=5, aspect=2)
#plot.set_axis_labels('duración', 'frecuencia')

plt.plot(x, pdf_gamma, 'r-', lw=2, label=f'Gamma fit (k={k:.2f}, θ={theta:.2f})')
plt.legend()

plt.show()
```

Ahora con el x_valor anterior de 821.68 segundos, miro para la función Gamma anterior, que probabilidad de cola habría (<span style="color: red;">Al final creo que esta probabilidad de contratación no es correcta </span>)

## 4. Pivotando el análisis después de comprender más los resultados

**Creo finalmente, que con las duraciones que hemos encontrado sabemos que por encima, la probabilidad de que no se contrate es practicamente nula** 

Después de hacer este análisis, otra cosa que se me ocurre es que tengo las duraciones de las llamadas del dataset y lo puedo dividir en rangos de tiempo. Por ejemplo llamadas entre 100-2000 segundos y ver la tasa de conversión de depósitos contratados. La tasa de conversión, sería como una estimación de probabilidad de éxito para estos rangos de tiempo.








Quiero trocear el segmento duración en trocitos más pequeños, pero para eso quiero ver los estadísticos princiaples

```python
df_BANK_marketing['duration'].describe()
```

```python
df_BANK_marketing['range_duration'] = pd.cut(df_BANK_marketing['duration'], bins=[x for x in range(2,3881,200)])
df_BANK_marketing
```

Quiero agrupar las etiquetas de range_duratino y sumar los 1 de cada rango de los depósitos y los registros por depósito

```python
agrup_duration_bins=df_BANK_marketing.groupby('range_duration')
serie_1=agrup_duration_bins['deposit'].sum()
serie_2=agrup_duration_bins['deposit'].count()

dataset_tasa_conversion = pd.concat([serie_1, serie_2], axis = 1)
dataset_tasa_conversion.columns=["sum_diposit_1","num_total_reg"]
dataset_tasa_conversion.reset_index(drop=False, inplace=True)
dataset_tasa_conversion
```

Calculo la tasa de conversion del anterior dataset

```python
dataset_tasa_conversion['tasa_conversion']=dataset_tasa_conversion['sum_diposit_1']/dataset_tasa_conversion['num_total_reg']
dataset_tasa_conversion = dataset_tasa_conversion.sort_values(by='tasa_conversion',ascending=False)
dataset_tasa_conversion
```

Visualizo la tasa de conversión vs los rangos

```python
#g=sns.barplot(x='range_duration',y='tasa_conversion',data=dataset_tasa_conversion)

g=sns.stripplot(x='range_duration', y='tasa_conversion', data=dataset_tasa_conversion)
#g=sns.catplot(x='range_duration', y='tasa_conversion', kind='bar', data=dataset_tasa_conversion, ci=None)
plt.xticks(rotation=90,fontsize=8)
ax=g.axes

plt.title('Gráfico de barras de range_duration vs_conversión')



Títol="Gráfic de barras de range_duration vs_tasa_conversión"
#g.set_titles("Gráfic de barras de range_duration vs_conversión")
#ax.set_title(Títol,fontweight='bold',color='darkblue')
ax.set_xlabel("range_duration [s]")
#ax.set_ylabel("freqüència")
#ax.grid(False)
# Mostrar el gráfico
plt.show()
```
