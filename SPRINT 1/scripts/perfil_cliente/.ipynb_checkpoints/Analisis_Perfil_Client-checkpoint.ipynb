```python
from sqlalchemy import create_engine, inspect
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import date, time, datetime

#Dades de connexió

usuario='EquipoB'

contraseña="E1q2u3i4p5oB"
host='212.227.90.6'
puerto=3306
nombre_base_datos='EquipB'

#Crear la URL de la connexió
url_conexion=f"mysql+pymysql://{usuario}:{contraseña}@{host}:{puerto}/{nombre_base_datos}"

try:
    #Crear el engine de SQLAlchemy
    engine=create_engine(url_conexion)
    cnx=engine.connect()

    print("conexió correcta")
    #cnx.close()
except Exception as e:
    print(f"Error al conectar con SQLAlchemy: {e}")
    
#Recordar que inspect del codi anterior quan creo la connexió amb sqlalchemy
#from sqlalchemy import create_engine, inspect
inspector=inspect(engine)
nombre_tablas=inspector.get_table_names()

# Diccionario para almacenar los DataFrames
df= {}

for taula in nombre_tablas:
    query=f"select * from {taula}"
    
    try:
        df[taula] = pd.read_sql(query, cnx)
        print(f"Taula '{taula}' carregada correctament.")
    except Exception as e:
        print(f"Error al carregar la taula '{taula}':{e}")
  
# Tancar la connexió
cnx.close()
```

```python
df_BANK_cliente=df['BANK_marketing']
```

```python
df_BANK_vista_cliente=df_BANK_cliente[['id','age','job','marital','education','balance','housing','loan','deposit']]
```

```python
# Crear grupos de edad (si no lo hemos hecho antes)
bins = [18, 30, 45, 60, 100]  
labels = ['Jóvenes (18-30)', 'Adultos jóvenes (31-45)', 'Adultos medios (46-60)', 'Mayores (>60)']
df_BANK_vista_cliente['age_group'] = pd.cut(df_BANK_vista_cliente['age'], bins=bins, labels=labels, right=False)

# Gráfico de la relación entre grupos de edad y productos financieros
for col in ['housing', 'loan', 'deposit']:
    plt.figure(figsize=(10,5))
    sns.countplot(x='age_group', hue=col, data=df_BANK_vista_cliente)
    plt.title(f"Relación entre grupos de edad y {col}")
    plt.show()

plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

# Codificar variables categóricas
df_encoded = df_BANK_vista_cliente.copy()

# Variables a codificar
categorical_cols = ['job', 'marital', 'education', 'housing', 'loan', 'age_group']

# Aplicar Label Encoding
for col in categorical_cols:
    df_encoded[col] = LabelEncoder().fit_transform(df_encoded[col])
```

```python
# Calcular la matriz de correlación
correlation_matrix = df_encoded.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración del tamaño de la figura
plt.figure(figsize=(12, 8))

# Heatmap de la matriz de correlación
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True, cbar=True)
plt.title('Matriz de Correlación')
plt.show()
```

```python
# Correlaciones específicas
correlation_with_housing = correlation_matrix['housing'].sort_values(ascending=False)
correlation_with_loan = correlation_matrix['loan'].sort_values(ascending=False)
correlation_with_deposit = correlation_matrix['deposit'].sort_values(ascending=False)

print("Correlación con Housing:")
print(correlation_with_housing)

print("\nCorrelación con Loan:")
print(correlation_with_loan)

print("\nCorrelación con Deposit:")
print(correlation_with_deposit)
```
